<u>Основи побудови графіків</u>
1. Створіть вектор розміром 11 зі значеннями від 0 до 5. Розрахуйте на його основі вектор y (y = x^2).

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

2. Побудова графіків.
Matplotlib підтримує два методи побудови графіків:
    - функціональний метод, який полягає у простій побудові графіків та їх
елементарному перетворенні.
    - об’єктно-орієнтований метод, який полягає у створенні графіків як
об’єктів та подальшому виклику методів та атрибутів цього об’єкта.

2.1. Функціональний метод. Побудова базового графіку, що
відображає залежність х від у.

```python
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

Побудова декількох графіків поруч з використанням команди
subplot:

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

2.2. Об’єктно-орієнтований метод. Етапи створення графіку:
    - ініціалізуйте створення графіку як об’єкта;
    - додайте до нього поле, для відображення графіка. В дужках задайте параметри графіку. Параметри задаються у  відсотковому значенні від 0 до 1 в наступному порядку: 
        1 – відстань від 0 по осі х 
        2 – відстань від 0 по осі у 
        3 – ширина поля, для відображення графіку 
        4 – висота поля, для відображення графіку
        При умові створення лише одного графіку на полі, параметри відповідають (0, 0, 1, 1)
    - додайте до поля графік та всі необхідні елементи.

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

Побудова декількох графіків з вказанням області відображення:

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.1, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'r')
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Title')

axes2.plot(y, x, 'b')
axes2.set_xlabel('X Label')
axes2.set_ylabel('Y Label')
axes2.set_title('Title')
```

Використання функції subplots при створенні графіка як об’єкта,
його ініціалізація:

```python
fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[1].plot(y, x)
```

```python
Використання каманди tight_layout() при частковому накладанні графіків.
```

```python
fig, axes = plt.subplots(nrows = 4, ncols = 4)
```

```python
fig, axes = plt.subplots(nrows = 4, ncols = 4)
plt.tight_layout()
```

Додавання легенди до графіку, використовуючи команду .legend()

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label = "x**2")
ax.plot(x, x**3, label = "x**3")
ax.legend()
```

Змінення розміру графіку та DPI.

```python
fig = plt.figure(figsize=(8,8), dpi=50)
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2,label="x**2")
ax.plot(x, x**3,label="x**3")
ax.legend()
```

Задання характеристик:

    - колір графіку

```python
fig, ax = plt.subplots()
ax.plot(x,x**2,color = 'b')
ax.plot(x,x**3,color = 'g')
```

    - тип та ширина лінії

```python
fig, ax = plt.subplots()
ax.plot(x,x**2,lw=2,ls = '--')
```

    - тип, розмір та колір маркерів

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, 
       marker = 'o',
       markersize = 10,
       markerfacecolor = 'r')
```

Збереження побудованого графіку (savefig).

```python
fig.savefig("filename.png")
```
